```python
n_news = []
n_comment = []
n_suppoter = []
t_purpose = []
t_term = []
t_percent = []
t_amount = []
n_like = []
n_img = []

driver = webdriver.Chrome("/Users/sgvin/chromedriver.exe")

for path in df1['url']:
    path = 'https://www.wadiz.kr' + path
    driver.get(path)
    time.sleep(0.5)
    soup2=BeautifulSoup(driver.page_source,"html.parser")
    time.sleep(0.5)
    # 새소식 / 커뮤니티 댓글 / 후원자 개수: 0인 경우 오류 벌생, 전체 텍스트 수집
    n_news.append(soup2.find_all('a', {'class':'tab-link'})[3].get_text()) # 새소식 개수
    n_comment.append(soup2.find_all('a', {'class':'tab-link'})[4].get_text()) # 커뮤니티 댓글 개수
    n_suppoter.append(soup2.find_all('a', {'class':'tab-link'})[5].get_text()) # 후원자 개수

    # 목표 금액 /  펀딩 기간
    t_purpose.append(soup2.find_all('p', {'style':'color:#00b2b2;font-size:13px;line-height:20px;margin-bottom:6px;'})[0].find_all('span')[0].get_text())
    t_term.append(soup2.find_all('p', {'style':'color:#00b2b2;font-size:13px;line-height:20px;margin-bottom:6px;'})[0].find_all('span')[1].get_text())

    # 달성률
    t_percent.append(soup2.find_all('p', {'style':'margin-bottom: 16px;'})[0].find('strong').get_text())
    
    # 달성금액
    t_amount.append(soup2.find('p', {'class':'total-amount'}).get_text())

    # 좋아요 개수
    n_like.append(soup2.find('em', {'class':'cnt-like'}).get_text())

    # 본문의 이미지 개수 
    n_img.append(len(soup2.find_all('img',{'class':'fr-fic fr-dib lazyload'})))
    
    driver.execute_script("window.open();")
    driver.switch_to.window(driver.window_handles[0])
    driver.close()
    driver.switch_to.window(driver.window_handles[-1])
# 시간이 많이 걸려 임시 중단
```

```python
df2 = pd.DataFrame({'news':n_news,
                 'comment': n_comment,
                 'suppoter':n_suppoter,
                 'purpose': t_purpose,
                    'term':t_term,
                   'percent':t_percent,
                   'amount':t_amount,
                   'like':n_like,
                   'img':n_img})
df2
```

```python
data = pd.read_excel('와디즈펀딩_0727ing.xlsx')
```

```python
# news, comment, suppoter에서 텍스트만 존재하는 경우 0값 입력
data['news'] = data['news'].apply(lambda x : x.split()[-1]).replace('새소식','0').astype(int)
data['comment'] = data['comment'].apply(lambda x : x.split()[-1]).replace('커뮤니티','0').astype(int)
data['suppoter'] = data['suppoter'].apply(lambda x : x.split()[-1]).replace('서포터','0').astype(int)
data
```

```python
import re
```

```python
data['purpose'] = data['purpose'].apply(lambda x: int(''.join(re.findall("\d+", x))))
data['amount'] = data['amount'].apply(lambda x: int(''.join(re.findall("\d+", x))))
data
```

```python
# 펀딩기간 -> 펀딩시작일, 펀딩마감일
data['start_date'] = data['term'].apply(lambda x : x.split('-')[0])
data['finsh_date'] = data['term'].apply(lambda x : x.split('-')[1])
data
```

```python
data.drop(['Unnamed: 0','term'], axis=1,inplace=True)
```
